```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

from shapely.geometry import Point
import matplotlib.pyplot as plt
from rasterio.mask import mask
from rasterio.plot import show
from rasterio.plot import plotting_extent
from natsort import natsorted
import earthpy.spatial as es
import pandas as pd
import geopandas
import rasterio
import pycrs
import os

file_dir = ('/Users/maddie/Projects/CPSC_597/data/data_raw/')
file_dir2 = ('/Users/maddie/Projects/CPSC_597/data/modified_data/')
```

```python
list_envirem_files=[]
names_envirem=[]

#add files to these lists
for root, dirs, files in os.walk(file_dir+"/bio_oracle/"):
    for file in files:
        if file.endswith('.tif'):
            list_envirem_files.append(file_dir+"/bio_oracle/"+file)
            name=file.replace(".tif","")
            names_envirem.append(name)
            names_envirem=natsorted(names_envirem,key=lambda y: y.lower())

            
            
list_gebco_files=[]
names_gebco=[] 
#add files to these lists
for root, dirs, files in os.walk(file_dir+'/gebcodata/'):
    for file in files:
        if file.endswith('.tif'):
            list_gebco_files.append(file_dir+'/gebcodata/'+file)
            name=file.replace(".tif","")
            names_gebco.append(name)
            names_gebco=natsorted(names_gebco,key=lambda y: y.lower())




list_envirem_future_files=[]
names_future_envirem=[]

#add files to these lists
for root, dirs, files in os.walk(file_dir+"/bio_oracle_future/"):
    for file in files:
        if file.endswith('.tif'):
            list_envirem_future_files.append(file_dir+"/bio_oracle_future/"+file)
            name=file.replace(".tif","")
            names_future_envirem.append(name)
            names_future_envirem=natsorted(names_future_envirem,key=lambda y: y.lower())

     
```

```python
for name in names_envirem:
    raster=rasterio.open(file_dir+"/bio_oracle/"+name+".tif")
    print(raster.meta) #inspect meta information
    show(raster,title="ENVIREM %s"%name)


for name in names_gebco:
    raster=rasterio.open(file_dir+"/gebcodata/"+name+".tif")
    print(raster.meta) #inspect meta information
    show(raster,title="GEBCO %s"%name)


for name in names_future_envirem:
    raster=rasterio.open(file_dir+"/bio_oracle_future/"+name+".tif")
    print(raster.meta) #inspect meta information
    show(raster,title="FUTURE ENVIREM %s"%name)
```

```python
list_envirem_bioclim_files=[]
for item in list_envirem_files:
    list_envirem_bioclim_files.append(item)


list_envirem_gebco_files = []
for item in list_gebco_files:
    list_envirem_gebco_files.append(item)

list_envirem_bioclim_future_files=[]
for item in list_envirem_future_files:
    list_envirem_bioclim_future_files.append(item)
    
names_envirem_bioclim=[]
for item in names_envirem:
    names_envirem_bioclim.append(item)

names_envirem_gebco = []
for item in names_gebco:
    names_envirem_gebco.append(item)

names_envirem_future_bioclim=[]
for item in names_future_envirem:
    names_envirem_future_bioclim.append(item)



with open(file_dir+'/bio_oracle/variable_list.txt', 'w') as f:
    for item in names_envirem_bioclim:
        f.write("%s\n" % item)      


with open(file_dir+'/gebcodata/variable_list.txt', 'w') as f:
    for item in names_envirem_gebco:
        f.write("%s\n" % item)      


with open(file_dir+'/bio_oracle_future/variable_list.txt', 'w') as f:
    for item in names_envirem_future_bioclim:
        f.write("%s\n" % item)      

 


es.stack(list_envirem_bioclim_files, file_dir2+'/stacked_bio_oracle/bio_oracle_stacked.tif')
es.stack(list_envirem_gebco_files, file_dir2+'/stacked_gebco/gebco_stacked.tif')
es.stack(list_envirem_bioclim_future_files, file_dir2+'/stacked_bio_oracle_future/bio_oracle_future_stacked.tif')
```
